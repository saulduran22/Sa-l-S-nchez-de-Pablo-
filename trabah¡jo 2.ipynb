```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
sheep_birth_rate = 1.1
sheep_death_rate = 1.05
wolf_birth_rate = 1.05
wolf_death_rate = 1.01
wolf_starvation_threshold = 5

# Initial conditions
initial_sheep_population = 80
initial_wolf_population = 50

# Simulation parameters
num_steps = 200

# Initialize populations
sheep_population = initial_sheep_population
wolf_population = initial_wolf_population

# Data collection
Sheeps_data = [sheep_population]
wolf_data = [wolf_population]

# Simulation loop
for _ in range(num_steps):
    # Lotka-Volterra equations
    sheep_births = sheep_birth_rate * sheep_population
    sheep_deaths = sheep_death_rate * sheep_population * wolf_population
    wolf_births = wolf_birth_rate * sheep_population * wolf_population
    wolf_deaths = wolf_death_rate * wolf_population

    # Update populations
    sheep_population = max(0, sheep_population + sheep_births - sheep_deaths)
    if sheep_population == 0:
        wolf_population = max(0, wolf_population - wolf_starvation_threshold)
    else:
        wolf_population = max(0, wolf_population + wolf_births - wolf_deaths)

    # Collect data
    Sheeps_data.append(sheep_population)
    wolf_data.append(wolf_population)

# Plotting the results
time_steps = np.arange(0, num_steps + 1)

plt.plot(time_steps, Sheeps_data, label='sheep')
plt.plot(time_steps, wolf_data, label='wolf')
plt.xlabel('Time Steps')
plt.ylabel('Population')
plt.title('wolf-sheep Ecosystem Simulation')
plt.legend()
plt.show()
```
